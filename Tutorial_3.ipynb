```python
import numpy as np

# 1. Create a numpy array of size 2 x 2

arr_2x2 = np.array([[1, 2],
                    [3, 4]])
print("--- Task 1: 2x2 Array ---")
print(arr_2x2)
print("\n")


# 2. Create a random numpy array of size 3 x 4

arr_random_3x4 = np.random.rand(3, 4)
print("--- Task 2: 3x4 Random Array ---")
print(arr_random_3x4)
print("\n")


# 3. Create a numpy array of ones and a numpy array of zeroes

arr_ones = np.ones((3, 3))
arr_zeros = np.zeros((3, 3))
print("--- Task 3: Ones and Zeros Arrays ---")
print("Array of ones:")
print(arr_ones)
print("Array of zeros:")
print(arr_zeros)
print("\n")


# 4. Create a random numpy array of size 6x3 and transpose it

arr_6x3 = np.random.rand(6, 3)
arr_transposed = arr_6x3.T
print("--- Task 4: Transpose ---")
print("Original 6x3 array:")
print(arr_6x3)
print("Transposed 3x6 array:")
print(arr_transposed)
print("\n")


# 5. Create two random numpy arrays of size n x p and p x m and compute the dot product

# n = 4
# p = 3
# m = 5
# So we need one array of (4, 3) and one of (3, 5).
n = 4
p = 3
m = 5

A = np.random.rand(n, p) # Creates a 4x3 array
B = np.random.rand(p, m) # Creates a 3x5 array



dot_product = np.dot(A, B)

print("--- Task 5: Dot Product ---")
print("Array A (shape 4x3):")
print(A)
print("Array B (shape 3x5):")
print(B)
print("Dot Product (shape 4x5):")
print(dot_product)
```
